```python
num_simulations = 10
# Modification pour stocker les magnétisations de chaque pas de temps pour chaque simulation
neutral_percents = [0, 0.01, 0.03, 0.1, 0.25, 0.3, 0.4, 0.5, 0.9]
monte_carlo_timeseries = {p: [] for p in neutral_percents}

# Boucle principale pour les simulations Monte Carlo
for simulation in range(num_simulations):
    print(f"Running simulation {simulation+1}/{num_simulations}")
    
    # Initialisation des résultats pour cette simulation
    magnetisation_values = {}
    full_grids = {}
    
    for neutral_percent in neutral_percents:
        alpha = 10
        SEED = 42 + simulation  # Différente graine pour chaque simulation
        np.random.seed(SEED)
        random.seed(SEED)
        reduced_alpha = -2.0 * beta * alpha
        
        spin_system = SpinSystem(
            grid_height=grid_height,
            grid_width=grid_width,
            init_up=init_up,
            fraction_neutral=neutral_percent,
            region_neutral="random")
        
        magnetisation = np.empty((min(total_updates, MAX_FILE_SIZE), ), dtype=float)
        magnetisation[:] = np.nan
        full_grid = []
        full_grid.append(reconstruct_grid(spin_system.black, spin_system.white))
        
        for ii in range(total_updates):
            mag = spin_system.update(reduced_neighbour_coupling, reduced_alpha)
            magnetisation[ii % MAX_FILE_SIZE] = mag
            full_grid.append(reconstruct_grid(spin_system.black, spin_system.white))
        
        magnetisation_values[neutral_percent] = magnetisation.copy()
        full_grids[neutral_percent] = full_grid.copy()
        
        # Stockage de la série temporelle complète pour cette simulation
        if simulation == 0:  # Initialisation des listes pour la première simulation
            monte_carlo_timeseries[neutral_percent] = [magnetisation.copy()]
        else:
            monte_carlo_timeseries[neutral_percent].append(magnetisation.copy())

# Calcul des statistiques Monte Carlo pour chaque pas de temps
mean_timeseries = {}
ci_lower_timeseries = {}
ci_upper_timeseries = {}

for neutral_percent in neutral_percents:
    # Convertir en tableau numpy pour faciliter les calculs
    all_simulations = np.array(monte_carlo_timeseries[neutral_percent])
    # Calcul de la moyenne pour chaque pas de temps
    mean_timeseries[neutral_percent] = np.mean(all_simulations, axis=0)
    
    # Calcul des intervalles de confiance pour chaque pas de temps
    ci_lower = np.zeros_like(mean_timeseries[neutral_percent])
    ci_upper = np.zeros_like(mean_timeseries[neutral_percent])
    
    for t in range(len(mean_timeseries[neutral_percent])):
        if not np.isnan(mean_timeseries[neutral_percent][t]):
            # Intervalles de confiance à 95% pour ce pas de temps
            t_values = all_simulations[:, t]
            t_values = t_values[~np.isnan(t_values)]  # Suppression des NaN
            if len(t_values) > 1:  # Vérifier qu'il y a suffisamment de données
                mean_val = np.mean(t_values)
                ci = stats.t.interval(0.95, len(t_values)-1, loc=mean_val, scale=stats.sem(t_values))
                ci_lower[t] = ci[0]
                ci_upper[t] = ci[1]
            else:
                ci_lower[t] = mean_timeseries[neutral_percent][t]
                ci_upper[t] = mean_timeseries[neutral_percent][t]
    
    ci_lower_timeseries[neutral_percent] = ci_lower
    ci_upper_timeseries[neutral_percent] = ci_upper

# Tracé des courbes de magnétisation avec moyenne et intervalles de confiance
colors = plt.cm.viridis(np.linspace(0, 1, len(neutral_percents)))
plt.figure(figsize=(12, 8))

for neutral_percent, color in zip(neutral_percents, colors):
    time_steps = np.arange(len(mean_timeseries[neutral_percent]))
    
    # Tracé de la moyenne
    plt.plot(time_steps, mean_timeseries[neutral_percent], 
             label=f'% of neutrals = {neutral_percent}', 
             color=color, linewidth=2)
    
    # Tracé des intervalles de confiance
    plt.fill_between(time_steps, 
                     ci_lower_timeseries[neutral_percent], 
                     ci_upper_timeseries[neutral_percent],
                     color=color, alpha=0.2)

plt.xlabel("Time Steps", fontsize=12)
plt.ylabel("Magnetization M(t)", fontsize=12)
plt.title("Impact of Neutral Agents on Market Magnetization\n(Mean with 95% CI across 10 simulations)", fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
